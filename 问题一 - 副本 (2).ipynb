############################################################计算事件恐怖主义分数

```python
import pandas as pd
input_file = open("df_use.csv",'rb')
df = pd.read_csv(input_file, header = 0,usecols=['eventid','nkill','nwound','propextent'])
df.head()
```

```python
df.describe()
```

```python
w_nkill=3
w_nwound=0.5
w_property=1
```

```python
danger_score_list=[]
scorelist=[]

xi_nkill=df['nkill'].max()-df['nkill'].min()
xi_nwound=df['nwound'].max()-df['nwound'].min()
xi_property=df['propextent'].max()-df['propextent'].min()
for i in range(len(df)):
    #df_nkill=df['nkill'][i]/100
    #df_nwound=df['dnwound'][i]/100
    #df_propextent=4-df['propextent'][i]
    df_nkill=(df['nkill'][i]-df['nkill'].min())/xi_nkill
    df_nwound=(df['nwound'][i]-df['nwound'].min())/xi_nwound
    df_propextent=(df['propextent'].max()-df['propextent'][i])/xi_property  #此处注意减法部分，与前两个不同
    id_score=[]
    id_score.append(df['eventid'][i])
    score=w_nkill*df_nkill+w_nwound*df_nwound+w_property*df_propextent
    id_score.append(score)
    danger_score_list.append(id_score)
    scorelist.append(score)
```

```python
import operator
sorted(danger_score_list, key = operator.itemgetter(1),reverse=True) 
```
